# I wanted to analyze the properties of those videos that were trending on the same day

1. Convert both trending data and publish time column to pandas date time format.
2. Find the difference in days between publish_time and trending_date
3. Find out those videos where the days difference is 0
4. Use those video to analyze the properties

```python
df.sort_values(by = "video_id")
```

```python
df.info()
```

```python
df['publish_time']-df['trending_date']
```

```python
df['trending_date']
```

https://strftime.org/

```python
df['trending_date'] = pd.to_datetime(df['trending_date'], format = '%y.%d.%m')
```

```python
df['trending_date']
```

```python
df.info()
```

```python
df['publish_time']
```

```python
df['publish_time'] = pd.to_datetime(df['publish_time'])
```

```python
df['publish_time']
```

```python
df.info()
```

```python
df['publish_time']-df['trending_date']
```

```python
df['publish_date'] = pd.to_datetime(df['publish_time'].dt.date)
```

```python
df.info()
```

```python
df['no_of_days_took_to_trend'] = (df['trending_date']-df['publish_date']).dt.days
```

```python
df.head()
```

```python
# How many videos are there which trended on the same day

df['no_of_days_took_to_trend'].value_counts()
```

```python
df_0 = df[df['no_of_days_took_to_trend']==0]
```

```python
df_0.head()
```

```python
df_0['video_id'].value_counts()
```

```python
df_0[df['video_id'] =="6ZGMomOaRqw"]
```

```python
df_0.drop_duplicates(subset = ['video_id', 'trending_date'], keep = 'last', inplace = True)
```

```python
df_0['video_id'].value_counts()
```

```python
df_0.head()
```

```python
# Most viewed video on the same trending date

df_0.sort_values(by = "views", ascending = False)
```

```python
# Channel wise same day trending video count

df_top_video = df_0.groupby("channel_title").agg({'video_id':'count'}).sort_values(by = 'video_id', ascending = False)
df_top_video.head(15)
```

```python
df_top_video = df_top_video.reset_index()
df_top_video
```
