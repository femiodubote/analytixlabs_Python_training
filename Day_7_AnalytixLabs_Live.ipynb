## How to filter data with Pandas

```python
df.Price < 10000
```

```python
df.Price.describe().round(3)
```

```python
df[df.Price < 100000]
```

```python
df.loc[df.Price < 100000, :]
```

```python
(df.Price <5000000) & (df.Rooms >3)
```

```python
df.loc[(df.Price <5000000) & (df.Rooms >3), ['Rooms', 'Type', 'Price']]
```

```python
df.loc[(df.Price <5000000) & (df.Rooms >3) | (df.Rooms>4), ['Rooms', 'Type', 'Price']]
```

```python
df.loc[df.Rooms ==3, ['Rooms', 'Type', 'Price']]
```

```python
df.loc[df.Rooms !=2, ['Rooms', 'Type', 'Price']]
```

```python
room = [8,12]

df.loc[df.Rooms.isin(room), :]
```
